```python
# Sol02:
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
X1 = [7.5, 5.8, 4.5, 3.7, 3.9, 5.7, 5.6, 5.9, 7.0, 6.3,3.3,3.3,
     6.2,5.1,6.9,5.8,4.3,5.9,6.9,6.5]
X2 = [8.6, 6.8, 7.4, 5.5, 5.6, 6.5, 6.9, 6.8, 8.2, 6.3,4.4,3.6, 6.4,5.1,6.8,8.0,5.1,5.8,8.4,7.0]
X3 = [5.9,7.1,5.9,7.8,9.1,6.8,7.4,7.4,8.5,10.2,5.8,4.3,8.9,8.0,9.1,8.4,6.0,9.0,6.8,6.4]

data = np.array([X1, X2, X3])
```

```python
avg= np.mean(data, axis=1)
```

```python
cov=np.cov(data, bias=True)
cov
```

```python
corr=np.corrcoef(data)
corr
```

```python
import matplotlib.pyplot as plt
plt.scatter(X1, X2, X3)
plt.show()
```

```python
import seaborn as sns
import statsmodels.api as sm
import scipy.stats as stats
```

```python
X1 = [3.3,3.3,3.7,3.9,4.3,4.5,5.1,5.6,5.7,5.8,5.8,5.9,5.9,6.2,6.3,6.5,6.9,6.9,7.0,7.5]
np_normal = pd.Series(X1) 
```
